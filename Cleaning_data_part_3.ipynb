```python
import numpy as np
import pandas as pd
pd.set_option("display.max_rows", None)
from ast import literal_eval
```

```python
jobs= pd.read_csv('jobs_fullycleaned.csv',index_col=0)
```

```python
jobs['how_old']= jobs['how_old'].str.split(' ').str[0]
```

```python
jobs['team_size'].value_counts()
```

```python
jobs['positives']= jobs['positives'].apply(literal_eval)
#convert to a list format
```

```python
all_factors = [i for sublist in jobs['positives'] for i in sublist ]

# Create a Pandas Series from the flattened list
factors_series = pd.Series(all_factors)
factors_series= factors_series.str.strip()
factors_series= factors_series[factors_series!= 'nan']
factors_series= factors_series.value_counts()
```

```python
factors_series
```

```python
factors_series.to_csv('positives.csv')
```

```python
jobs.head()
```

```python
def listit(stringg):
  negatives=[]
  for i in stringg.split(','):
    negatives.append(i)
  return negatives
```

```python
jobs['negatives']= jobs['negatives'].astype(str).apply(listit)
```

```python
all_factors = [i for sublist in jobs['negatives'] for i in sublist ]

# Create a Pandas Series from the flattened list
factors_series = pd.Series(all_factors)
factors_series= factors_series.str.strip()
factors_series= factors_series[factors_series!= 'nan']
factors_series= factors_series.value_counts()
```

```python
jobs= jobs.drop(columns=['Unnamed: 0'])
```
