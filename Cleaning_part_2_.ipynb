```python
import pandas as pd
import numpy as np
pd.set_option("display.max_rows", None)
```

```python
jobs= pd.read_csv('jobs_cleaned.csv',index_col=0)
jobs.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
jobs.head()
```

```python
jobs['avg_salary']=jobs['avg_salary'].replace('--',np.nan)
```

```python

def salary(amount):
    if isinstance(amount, str) and 'k' in amount:
        a = amount.replace('k', '')
        return float(a) * 1000
    else:
        return amount

jobs['avg_salary'] = jobs['avg_salary'].fillna(0).apply(salary)

jobs['avg_salary']=jobs['avg_salary'].astype(int)
```

```python
jobs.head()
```

```python
jobs['jobs'] = jobs['jobs'].apply(salary)
jobs['jobs']=jobs['jobs'].fillna(0).astype(int,errors='ignore')
```

```python
jobs= jobs.replace('--',np.nan)
```

```python
jobs['benefits'] = jobs['benefits'].apply(salary)
jobs['benefits']=jobs['benefits'].fillna(0).astype(int)
```

```python
jobs.head()
```

```python
jobs['team_size'].value_counts()
```

```python
def employee(data):
  if '-' in data:
    return int(data.split('-')[0].split('k')[0])
  elif isinstance(data, str) and 'Lakh+' in data:
    b= data.split(' Lakh+')[0]
    return int(b)* 100000
  else:
    return np.nan

a= '1 Lakh+'
employee(a)
```

```python
jobs['city'].value_counts()
```

```python
jobs['city']=jobs['city'].replace('Bengaluru/Bangalore','Bengaluru')
jobs['city']=jobs['city'].replace('Delhi','New Delhi')
jobs['city']=jobs['city'].replace('Navi','Navi Mumbai')
```

```python
def city(name):
 return name.split('+')[0].strip().capitalize()

```

```python
jobs['city']= jobs['city'].astype(str)
jobs['city']= jobs['city'].apply(city)
```

```python
def positives(lists):
  a= lists.split(',')
  return a
```

```python
jobs['positives']= jobs['positives'].astype(str)
jobs['positives']= jobs['positives'].apply(positives)
jobs['negatives']= jobs['negatives'].astype(str)
jobs['negatives']= jobs['negatives'].apply(positives)
```
