```python
import numpy as np
import pandas as pd
pd.set_option("display.max_rows", None)
```

```python
jobs= pd.read_csv('jobs.csv',index_col=0)
```

```python
jobs.head()
```

```python
jobs['state'].value_counts().head(10
                   )
```

```python
new_order= [	'name',	'rating',	'avg_salary',	'jobs'	,'benefits',	'type_of_company',	'team_size' ,'state'	,'how_old'	,'city',		'positives',	'negatives']
jobs= jobs[new_order]
jobs.head()
```

```python
temp_clean= jobs[jobs['type_of_company'].str.contains('employees', case=False, na=False)]
temp_clean['type_of_company'].value_counts().head(10)
```

```python
temp_clean.head()
```

```python

columnsshift= ['type_of_company', 'team_size'	,'state'	,'how_old',	'city']
x = temp_clean[columnsshift].shift(1, axis=1)
temp_clean[columnsshift]= x
temp_clean.shape
```

```python
temp_clean2= temp_clean[~temp_clean['state'].isin(['Public'])]
temp_clean2.shape
```

```python
columnsshift= ['state'	,'how_old',	'city']
x = temp_clean2[columnsshift].shift(1, axis=1)
temp_clean2[columnsshift]= x
temp_clean2
```

```python
temp_clean2['how_old'].value_counts()
```

```python
temp_clean3= temp_clean2[temp_clean2['how_old'].str.endswith('more')]
columnsshift= ['how_old',	'city']
x = temp_clean3[columnsshift].shift(1, axis=1)
temp_clean3[columnsshift]= x
temp_clean3
```

```python
temp_clean2 = temp_clean2.dropna(subset=['how_old'])
temp_clean2.loc[temp_clean2['how_old'].str.endswith('more')]
```

```python
temp_clean[~temp_clean['state'].isin(['Public'])] = temp_clean2[~temp_clean2['state'].isin(['Public'])]
```

```python

a= jobs[jobs['type_of_company'].str.endswith('more',na=False)]
a['city'] =a['type_of_company']
a['type_of_company']= np.nan
jobs.loc[jobs['type_of_company'].str.endswith('more',na=False)]=a
```

```python
jobs['type_of_company'].value_counts()
```

```python
jobs.to_csv('job_safe.csv')
```
